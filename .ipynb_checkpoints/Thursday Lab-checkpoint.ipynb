```python
df.columns = ['sex', 'rank', 'year', 'degree', 'y_deg', 'salary']
```

```python
# Create dummy variables for each categorical variables.

pd.get_dummies(df)

df1 = pd.concat([df, df['sex'].str.get_dummies(sep=',')], axis=1)
df2 = pd.concat([df1, df1['rank'].str.get_dummies(sep=',')], axis=1)
df3 = pd.concat([df2, df2['degree'].str.get_dummies(sep=',')], axis=1)
df3
```

```python
df3.groupby('sex').salary.agg(['count', 'min', 'max', 'mean'])
```

```python
df3.groupby('rank').salary.agg(['count', 'min', 'max', 'mean'])
```

```python
df3.groupby('degree').salary.agg(['count', 'min', 'max', 'mean'])
```

```python
dfm = df3[(df3['male']== 1 )]
# we already sorted for Male so every one of these is male
```

```python
dfm = dfm[['salary','year','y_deg', 'assistant','associate','full','doctorate','masters']]
dfm.head()
```

```python
y = dfm.iloc[:,0:1]
X = dfm.iloc[:,1:]
X.head()
```

```python
y.head()
```

```python
import numpy as np
import pandas as pd
from scipy import stats
from sklearn import datasets
import statsmodels.api as sm
import patsy
```

```python
model = sm.OLS(y, X).fit()

predictions = model.predict(X)
print "r^2:", model.rsquared_adj

# Plot the data and the best fit line
## The data
# plt.scatter(X, y)
# ## The line / model
# plt.plot(X, predictions)
# plt.ylabel("Salary")
# plt.xlabel('year, y_deg, female, assistant, associate, full, doctorate, masters')
model.summary()
```

```python
rlm = linear_model.Ridge(alpha=0, normalize=True)

# Fit the polynomial again with ridge regularization
ridge_model = rlm.fit(X, y)
predictions = ridge_model.predict(X)

# plt.scatter(X, y)
# plt.title("Randomly Generated Data")
# plt.plot(X, predictions)
# plt.show()
ridge_model.score(X, y)

#this is my model coeficients
ridge_model.coef_
```
